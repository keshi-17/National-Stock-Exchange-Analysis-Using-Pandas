# National Stock Exchange Analysis Using Pandas

## Objective:
The primary objective is to conduct a comprehensive analysis of stocks listed in the official NSE indices, including Nifty50, NiftyNext50, NiftyMidcap150, NiftyIT, NiftyBanks, NiftyAuto, NiftyFMCG, and NiftyPharma. The analysis will be executed using the pandas library, focusing on extracting meaningful insights and trends.

Disclaimer: This analysis is intended solely for educational purposes and serves as a practical application of pandas for real-world financial data. It is not intended as financial advice, and any investment decisions should be made after thorough research and consultation with financial experts.ts.

## Dataset Overview: 
Relevant stock information was extracted from the Yahoo Finance website for this analysis using the Python module "yahoofinancials." "net income including noncontrolling interests," which is each company's equivalent of profit after taxes, is the important indicator that is taken into account.


Although the original idea was to use ten years of historical data, practical considerations forced a more condensed study that concentrated on the last three years of data. A meaningful picture of the financial performance of the chosen equities within the designated indices is given by this timeframe.
